```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
gps_data = pd.read_csv("gps-data.csv")
gps_data.head()
```

```python
gps_data.dtypes
```

```python
# Convertimos los datos del gps a HH.hh y los agregamos en una nueva columna
gps_data["Latitud[HH.hh]"] = (gps_data["Latitud"]//100)+((gps_data["Latitud"]%100)/60)
gps_data["Longitud[HH.hh]"] = (gps_data["Longitud"]//100)+((gps_data["Longitud"]%100)/60)

# Convertimos los datos del gps a HH°mm'ss.ss'' y los agregamos a una nueva columna
gps_data["Latitud[HH°mm'ss'']"] = [f"{round(i//100)}°{math.floor(i%100)}'{round((i%1)*60,4)}''" for i in gps_data["Latitud"]]
gps_data["Longitud[HH°mm'ss'']"] = [f"{round(i//100)}°{math.floor(i%100)}'{round((i%1)*60,4)}''" for i in gps_data["Longitud"]]
gps_data
```

```python
# Coordenada tomada con el telefono gps
punto = (20.11302,98.845231)
```

```python
from geopy.distance import geodesic
```

```python
plt.scatter(gps_data["Longitud[HH.hh]"], gps_data["Latitud[HH.hh]"], label="Puntos Obtenidos")
plt.scatter(punto[1], punto[0], color='red', label="Coordenada real")
# Añadir leyenda
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
gps_data["Error Absoluto Latitud"] = abs(gps_data["Latitud[HH.hh]"] - punto[0])
gps_data["Error Absoluto Longitud"] = abs(gps_data["Longitud[HH.hh]"] - punto[1])
gps_data
```

```python
error_promedio_latitud = gps_data["Error Absoluto Latitud"].mean()
error_promedio_latitud
```

```python
error_promedio_longitud = gps_data["Error Absoluto Longitud"].mean()
error_promedio_longitud
```

```python
puntos_obtenidos = list(zip(gps_data["Latitud[HH.hh]"], gps_data["Longitud[HH.hh]"]))
# puntos_obtenidos
```

```python
gps_data["Diferencia[m]"] = [geodesic(punto, i).meters for i in puntos_obtenidos]
gps_data
```

```python
print(f"Radio de Error[m] = {gps_data["Diferencia[m]"].max()}m")
```
